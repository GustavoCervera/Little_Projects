```python
#=================================================
#Programa inicial sobre lista de contatos
#Gustavo Zacarias Cervera, feito em 16-09-2020
#=================================================

import json

#Carrega o dicionario ja salvo como um banco de dados
dicionario = json.load(open("Projetinho_Lista_de_Contatos_TXT.txt"))
#Looping para geracao da interface do usuario
while True:
#Cabecalho da interface do usuario
    print("\n")
    print("="*23+"="*24)
    print("="*14+" LISTA DE CONTATOS "+"="*14)
    print("="*23+"="*24)
    print("\n\tL - Visualizar a lista de contatos")
    print("\tI - Visualizar as informacoes de um contato")
    print("\tA - Adicionar um contato")
    print("\tR - Remover um contato")
    print("\tS - Sair da aplicação")
    print("\n")
    print("="*23+"="*24)
    print("="*23+"="*24)
    print("="*23+"="*24)
#Entrada da resposta do que o usuario pretende fazer
    entry = input("\nO que deseja fazer? ")
    
#Mostra ao usuario a lista de contatos
    if entry in ["L","l"]:
        print("\n")
        print(dicionario)
        continue

#Adiciona um novo contato a lista de contatos
    elif entry in ["A", "a"]:
        dados = []
        name = input("\n\t Digite o nome que deseja adicionar: \n")
        if name in dicionario:
            print("\nEsse nome ja existe")
        else:
            number = input("\n\t Agora digite o numero de telefone: \n")
            dados.append(number)
            email = input("\n\t Agora digite o email: \n")
            dados.append(email)
            sexo = input("\n\t Agora digite o sexo: \n")
            dados.append(sexo)
            dicionario[name] = dados
            
#Visualiza as informacoes de um contato
    elif entry in ["I", "i"]:
        name_i = input("\n\t Qual o contato que voce deseja saber as informacoes: ")
        if name_i not in dicionario:
            print("\nO nome digitado nao consta no cadastro\n")
        else:
            lista_i = dicionario[name_i]
            info_name = name_i
            info_tel = lista_i[0]
            info_email = lista_i[1]
            info_sexo = lista_i[2]
            print("\n\n")
            print("O usuario {} tem os seguintes dados: ".format(info_name))
            print("\nTelefone: {}".format(info_tel))
            print("\nE-mail: {}".format(info_email))
            print("\nsexo: {}".format(info_sexo))

#Remove o contato desejado da lista de contatos
    elif entry in ["R", "r"]:
        name_d = input("\n\t Digite o nome que deseja apagar: \n")
        if name_d not in dicionario:
            print("\nO nome digitado nao consta no cadastro\n")
        else:
            del dicionario[name_d]
            print("\nO contato {} foi deletado da sua lista.\n".format(name_d))
            
#Encerra a aplicacao
    elif entry in ["S", "s"]:
        print("\nObrigado por usar a lista de contatos\n")
        #Antes de encerrar a aplicacao, o programa salva a atual lista de contatos em um database
        json.dump(dicionario, open("Projetinho_Lista_de_Contatos_TXT.txt","w"))
        break
        
#Indica ao usuario que algum valor errado foi inserido
    else:
        print("\nPor favor, entre com um valor valido.\n")
        continue
```
